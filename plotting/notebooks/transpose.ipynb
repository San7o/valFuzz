```python
# libraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from unum import Unum
# data
from data.data import *

# setup
sns.set_theme(style="whitegrid", palette="Set2")
# sns.set_theme(style="ticks", palette="Set2")
# Read the data
raw_data: pd.DataFrame = data.parse_data('../reports/base.txt')
```

# Transpose

```python
df_base_transpose: pd.DataFrame = raw_data.loc[raw_data.name == 'matrix transpose base' ].copy()
print(df_base_transpose)

sns.regplot(  # linear Regression
    data=df_base_transpose, label="matrix transpose base",
    x="space", y="time",
    order=1, ci=None
)
plt.title('Time over input size')
# Whould I be using a logarictimic scale because the value range is large?
# plt.yscale('log')
# plt.xscale('log')
plt.xlabel('Input Size (bytes)')
plt.ylabel('Time (s)')
plt.legend()
plt.show()
```
